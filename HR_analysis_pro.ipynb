```
#try
# FEATURE SELECTION
#Automatic implementations of Backward Elimination in Python with p-value
import statsmodels.api as sm
def backwardElimination(y,x,sl,x_test):
    numVars = len(x[0])
    numRows = len(x)
    x = np.append(arr = np.ones( (numRows,1) ).astype(int), values=x, axis=1)
    x_test = np.append(arr = np.ones( (len(x_test),1) ).astype(int), values=x_test, axis=1)
    for i in range(0, numVars):
        print(i,end=' ')
        numVars = len(x[0])
        regressor_OLS = sm.OLS(y, x).fit()
        pVal=regressor_OLS.pvalues.astype(float)
        maxVar = max(pVal)
        if maxVar > sl:
            for j in range(0, numVars):
                if (pVal[j] == maxVar):
                    x = np.delete(x, j, 1)
                    x_test = np.delete(x_test, j, 1)
                    break;
        else:
            break
    print('sup')
    regressor_OLS.summary()
    return x,x_test
x_opt,x_test=backwardElimination(y_res,X_res,0.05,PC_X_train_B)
```

```
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_opt, y_res, test_size = 0.2, random_state = 0)

```
